REVERSE OF VOWELS IN A STRING

```python
class Solution(object):
    def reverseVowels(self, s):
        self.s=s
        v="AEIOUaeiou"
        l=list(s)
        a=0
        b=len(s)-1
        while a<b:
          while l[a] not in v and a<b:
            a+=1
          while l[b] not in v and a<b:
            b-=1
          l[a],l[b]=l[b],l[a]
          a+=1
          b-=1
        s1=''.join(l)
        print(s1)
c=Solution()
c.reverseVowels(input("enter string :"))
```
